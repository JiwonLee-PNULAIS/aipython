```python
plt.rc("font", family="NanumGothic")

df4 = pd.DataFrame({"도시":["서울", "부산", "인천", "대구"], "인구 수":[9904312, 3448737, 2890451, 2466052]})

plt.bar(df4["도시"], df4["인구 수"])
plt.title("도시별 인구 수")
plt.xlabel("도시")
plt.ylabel("인구 수")
plt.show()
```
