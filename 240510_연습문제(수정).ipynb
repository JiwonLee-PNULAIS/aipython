```python
df_kor = df[df["대표국적"]=="한국"]
df_usa = df[df["대표국적"]=="미국"]
df_uk = df[df["대표국적"]=="영국"]
df_jpn = df[df["대표국적"]=="일본"]
```

```python
print("한국의 최상위 순위: ", df_kor["순위"].min())
print("미국의 최상위 순위: ", df_usa["순위"].min())
print("영국의 최상위 순위: ", df_uk["순위"].min())
print("일본의 최상위 순위: ", df_jpn["순위"].min())
```
