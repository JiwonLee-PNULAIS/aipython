```python
import matplotlib.pyplot as plt
import pandas as pd

df3 = pd.DataFrame({"요일":["월", "화", "수", "목", "금", "토", "일"], "공부시간(시간)":[2, 3, 1, 5, 4, 2, 1], "성적":[70, 75, 65, 85, 80, 70, 60]})
df3
```

```python
plt.plot(df3["요일"], df3["공부시간(시간)"], marker="o", label="요일별 공부 시간")
plt.show()
```

```python
plt.plot(df3["요일"], df3["성적"], marker="^", label="요일별 성적")
plt.show()
```
